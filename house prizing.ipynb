```python
plt.figure(figsize=(8, 4))
plt.hist(df['SalePrice'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of sels Prices')
plt.xlabel('SalePrice')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.kdeplot(df['SalePrice'], fill=True, color='orange')
plt.title('Kernel Density sels Prices')
plt.xlabel('Price')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(8, 2))
sns.boxplot(x=df['SalePrice'], color='lightgreen')
plt.title('Boxplot of House Prices')
plt.xlabel('Price')
plt.show()
```

```python
corr_matrix = df.corr(numeric_only=True)
print("Correlation Matrix:")
print(corr_matrix)
```

```python
plt.figure(figsize=(30, 24))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Replace column names with your actual dataset's column names if different
key_features = ['BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath', 'BedroomAbvGr','KitchenAbvGr']
target = 'SalePrice'

# Pairplot to visualize relationships
sns.pairplot(df[[*key_features, target]])
plt.suptitle('Relationships between Key Features and House Price', y=1.02)
plt.show()
```

```python
# Scatter plots for each feature vs price
for feature in key_features:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=df[feature], y=df[target])
    plt.title(f'{feature.capitalize()} vs Price')
    plt.xlabel(feature.capitalize())
    plt.ylabel('Price')
    plt.show()
```
